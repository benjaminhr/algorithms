```python
class ArrayList:
    def __init__(self):
        self.inArray = [0 for i in range(10)]
        self.count = 0
        
    def swap(self, i, j):
        tmp = self.inArray[i]
        self.inArray[i] = self.inArray[j]
        self.inArray[j] = tmp
    
    def get(self, i):
        return self.inArray[i]

    def set(self, i, e):
        self.inArray[i] = e

    def length(self):
        return self.count

    def append(self, e):
        self.inArray[self.count] = e
        self.count += 1
        if len(self.inArray) == self.count:
            self._resizeUp()     # resize array if reached capacity

    def insert(self, i, e):
        for j in range(self.count,i,-1):
            self.inArray[j] = self.inArray[j-1]
        self.inArray[i] = e
        self.count += 1
        if len(self.inArray) == self.count:
            self._resizeUp()     # resize array if reached capacity

    def remove(self, i):
        self.count -= 1
        val = self.inArray[i]
        for j in range(i,self.count):
            self.inArray[j] = self.inArray[j+1]
        return val
    
    def __str__(self):
        return str(self.inArray[:self.count])

    def _resizeUp(self):
        newArray = [0 for i in range(2*len(self.inArray))]
        for j in range(len(self.inArray)):
            newArray[j] = self.inArray[j]
        self.inArray = newArray
```

```python
class Heap:
    def __init__(self):
        self.inList = ArrayList()
        self.size = 0
        
    def __str__(self):
        return str(self.inList)
        
    def add(self, d):
        self.inList.append(d)
        pos = self.size
        self.size += 1
        while pos > 0 and self.inList.get(pos) > self.inList.get((pos-1)//2):
            self.inList.swap(pos,(pos-1)//2)
            pos = (pos-1)//2
        
    def heapify(self, pos):
        if 2*pos+1 >= self.size: return
        if 2*pos+2 >= self.size or self.inList.get(2*pos+1) > self.inList.get(2*pos+2):
            maxChild = 2*pos+1
        else: maxChild = 2*pos+2
        if self.inList.get(maxChild) < self.inList.get(pos): return
        self.inList.swap(pos,maxChild)
        self.heapify(maxChild)
        
    def removeRoot(self):
        val = self.inList.get(0)
        self.inList.set(0,self.inList.get(self.size-1))
        self.inList.remove(self.size-1)
        self.size -= 1
        self.heapify(0)
        return val
    
    def append(self, d):
        self.inList.append(d)
        self.size += 1
        
    def max(self): 
        return self.inList.get(0)
    
    def getRelative(self, pos, flag):
        index = None
        if flag == 0: index = (pos - 1) // 2
        elif flag == 1: index = 2*pos+1
        elif flag == 2: index = 2*pos+2
        
        if index and index <= self.size: 
            return self.inList.get(index)
    
    def sumAll(self):
        sum = 0
        for i in range(self.size):
            sum += self.inList.get(i)
        return sum
    
    def min(self):
        last = self.inList.get(self.size-1)
        minimum = last
        
        for i in range((last - 1)//2, -1, -1):
            curr = self.inList.get(i)
            leftChild = 2*curr+1
            rightChild = 2*curr+2
            
            if leftChild < self.size and self.inList.get(leftChild) < minimum:
                minimum = self.inList.get(leftChild)
            elif rightChild < self.size and self.inList.get(rightChild) < minimum:
                minimum = self.inList.get(rightChild)
        return minimum
    
    def removeVal(self, d):
        index = None
        for i in range(self.size):
            curr = self.inList.get(i)
            if curr == d:
                index = i 
        
        if index is None: return
        val = self.inList.remove(index)
        for i in range(self.size):
            self.heapify(i)
        return val
    
    def heapSort1(self, A):
        h = Heap()
        for i in range(len(A)):
            h.add(A[i])
        for i in range(len(A)-1,-1,-1):
            A[i] = h.removeRoot()
        return A
    
    def toSortedArray(self):
        A = [0 for i in range(self.size)]
        for i in range(self.size):
            A[i] = self.inList.get(i)
        return self.heapSort1(A)
        
def heapify(A, length, i):
    largest = i
    leftChild = 2*i+1
    rightChild = 2*i+2
    
    if leftChild < length and A[i] < A[leftChild]:
        largest = leftChild
    
    if rightChild < length and A[largest] < A[rightChild]:
        largest = rightChild
    
    if largest != i:
        A[i], A[largest] = A[largest], A[i]
        heapify(A, length, largest)
    
def heapSort3(A):
    length = len(A)
    for i in range(length, -1, -1):
        heapify(A, length, i)
    
    for i in range(length-1, 0, -1):
        A[i], A[0] = A[0], A[i]
        heapify(A, i, 0)
    return A
    
H = Heap()
H.add(10); H.add(9); H.add(20); H.add(3); H.add(17); H.add(2);
H.add(21); H.add(4); H.add(8); H.add(4); H.add(3); H.add(10);

A = [5,6,2,7,1,2]
B = heapSort3(A)
#print(B)
```

```python
class PQElement:
    def __init__(self, val, p):
        self.val = val
        self.p = p
    
    def __lt__(self, other):
        return self.p < other.p
    
    def __gt__(self, other):
        return self.p > other.p
    
    def __str__(self):
        return "(" + str(self.val) + ", " + str(self.p) + ")"

class PQueue:
    def __init__(self):
        self.heap = Heap()
        self.size = 0
    
    def size(self):
        return self.size
    
    def enq(self, element):
        self.size += 1
        self.heap.add(element)
        
    def deq(self):
        self.size -= 1
        return self.heap.removeRoot()
    
    def __str__(self):
        arr = self.heap.toSortedArray()
        res = "["
        for i in range(len(arr)):
            res += str(arr[i])
        return res + "]"
            
```

```python
pq = PQueue()
for i in range(15):
    pq.enq(PQElement(i,10-i))
    
print(pq)
print(pq.deq(),pq)
```
