```python
class ArrayList:
    def __init__(self):
        self.inArray = [0 for i in range(10)]
        self.count = 0
        
    def get(self, i):
        try:
            self._checkBounds(i, len(self.inArray))
            return self.inArray[i]
        except Exception as e:
            print(e)

    def set(self, i, e):
        try:
            self._checkBounds(i, len(self.inArray))
            self.inArray[i] = e
        except Exception as e:
            print(e)  

    def length(self):
        return self.count        

    def append(self, e):
        self.inArray[self.count] = e
        self.count += 1
        if len(self.inArray) == self.count:
            self._resizeUp()
    
    def appendAll(self, A):
        temp = [0 for i in range(len(A) + self.count + 10)]
        
        for i in range(self.count):
            temp[i] = self.inArray[i]
        
        for i in range(self.count, len(A)):
            temp[i] = A[i]
        
        self.inArray = temp

    def insert(self, i, e):
        try:
            self._checkBounds(i, len(self.inArray))
            for j in range(self.count,i,-1):
                self.inArray[j] = self.inArray[j-1]     
            self.inArray[i] = e
            self.count += 1
            if len(self.inArray) == self.count:
                self._resizeUp()
                
        except Exception as e:
            print(e)
    
    def remove(self, i):
        try:
            self._checkBounds(i, len(self.inArray))
            val = self.inArray[i]
            for j in range(i, self.count):
                self.inArray[j] = self.inArray[j+1]

            self.count -= 1
            return val
        
        except Exception as e:
            print(e)

    def removeVal(self, e):        
        for i in range(self.count):
            if self.inArray[i] == e:
                self.remove(i)
                return True
        return False

    def _resizeUp(self):
        newArray = [0 for i in range(2*len(self.inArray))]
        for j in range(len(self.inArray)):
            newArray[j] = self.inArray[j]
        self.inArray = newArray
        
    def _checkBounds(self, i, hi):  # checks whether i is in [0,hi]
        if i < 0 or i > hi:
            raise Exception("index "+str(i)+" out of bounds!")
    
    def clone(self):
        newList = ArrayList()
        for i in range(self.count):
            el = self.inArray[i]
            newList.append(el)
            
        return newList
    
    def toArray(self):
        arr = [self.inArray[i] for i in range(self.count)]
        return arr
    
    def _insertionSort(self, k, A, hi):
        for i in range(hi - 1, -1, -1):
            if k >= A[i]:
                A[i + 1] = k
                return
            
            A[i + 1] = A[i]            
        A[0] = k
    
    def sort(self):
        A = self.inArray
        
        for i in range(1, self.count):
            self._insertionSort(A[i], A, i)
        

A = ArrayList()
A.appendAll([5,6,3,1,8,7,2])
A.sort()
print(A.inArray)
```

```python
"""
Write a Python class PQueue that implements a priority queue using an array list of
PQElement’s. In particular, you need to implement 5 functions:
• one for creating an empty priority queue
• one for returning the size of the priority queue
• one for enqueueing a new PQ-element in the priority queue
• one for dequeueing from the priority queue the PQ-element with the greatest priority
• one that prints the elements of the priority queue into a string (call this one __str__)
"""
class PQElement:
    def __init__(self, v, p):
        self.val = v
        self.priority = p
        
    def __str__(self):
        return "("+str(self.val)+","+str(self.priority)+")"
    
    
class PQueue:
    def __init__(self):
        self.queue = ArrayList()
    
    def size(self):
        return self.queue.length()
    
    def enq(self, element):
        qLength = self.queue.length()
        
        if qLength == 0:
            self.queue.insert(0, element)
            return
        
        for i in range(0, qLength):
            curr = self.queue.get(i)
            
            if curr.priority > element.priority:
                self.queue.insert(i, element)
                return
            
        self.queue.insert(qLength, element)
            
    def deq(self):
        qLength = self.queue.length()
        return self.queue.remove(qLength - 1)
    
    def __str__(self):
        arr = self.queue.toArray()
        string = ""
            
        for i in range(0, len(arr)):
            string += str(arr[i])
        
        return string
            
ls = PQueue()
for i in range(15):
    ls.enq(PQElement(i,10-i))
print(ls)
print(ls.deq(),ls)
```
