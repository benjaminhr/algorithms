```python
class Node:
    def __init__(self, d, n):
        self.data = d
        self.next = n

class LinkedList:
    def __init__(self):
        self.head = None
        self.length = 0

    def __str__(self):
        st = "--> "
        ptr = self.head
        while ptr != None:
            st = st + str(ptr.data)
            st = st+" -> "
            ptr = ptr.next
        return st+"None"
        
    def last(self):
        if self.head == None:
            return None
        ptr = self.head
        while ptr.next != None:
            ptr = ptr.next
        return ptr.data
        
    def lastNode(self):
        if self.head == None:
            return None
        ptr = self.head
        while ptr.next != None:
            ptr = ptr.next
        return ptr

    def search(self, d):
        i = 0
        ptr = self.head
        while ptr != None:
            if ptr.data == d:
                return i
            ptr = ptr.next
            i += 1
        return -1
        
    def append(self, d):
        if self.head == None:      
            self.head = Node(d,None) 
        else:
            ptr = self.head
            while ptr.next != None:
                ptr = ptr.next
            ptr.next = Node(d,None)
        self.length += 1
        
    def appendNode(self, node):
        if self.head == None:      
            self.head = node
        else:
            ptr = self.head
            while ptr.next != None:
                ptr = ptr.next
            ptr.next = node
        self.length += 1

    def insert(self, i, d):
        if self.head == None or i == 0:
            self.head = Node(d,self.head)
        else:
            ptr = self.head
            while i>1 and ptr.next != None:
                ptr = ptr.next
                i -= 1
            ptr.next = Node(d,ptr.next)
        self.length += 1

    def remove(self, i): # removes i-th element and returns it
        if self.head == None:
            return None
        if i == 0:
            val = self.head.data
            self.head = self.head.next
            self.length -= 1
            return val
        ptr = self.head
        while i>1 and ptr.next != None:
            ptr = ptr.next
            i -= 1
        if i == 1:
            val = ptr.next.data
            ptr.next = ptr.next.next
            self.length -= 1
            return val
    
    def removeVal(self, d):
        if self.head == None:
            return
        if self.head.data == d:
            self.head = self.head.next
            self.length -= 1
        else:
            ptr = self.head	
            while ptr.next != None:
                if ptr.next.data == d:
                    ptr.next = ptr.next.next
                    self.length -= 1
                    break
                ptr = ptr.next
    
    def sublist(self, i):
        ptr = self.head
        ls = LinkedList()
        ls.length = self.length
        while ptr != None and i>0:
            ptr = ptr.next
            i -= 1
            ls.length -= 1
        ls.head = ptr
        return ls
    
    def appendAll(self, A):
        for i in range(len(A)):
            self.append(A[i])
            
    def nullify(self): 
        self.head = None 
        
    def merge(self, other):
        listOneNode = self.lastNode() 
        listTwoNode = other.head
        listOneNode.next = listTwoNode            
        
    def sort(self):
        smaller = LinkedList()
        other = LinkedList()
        
        if self.head == None:
            return
        
        pivot = self.head.data
        ptr = self.head.next
        
        while ptr != None:
            if ptr.data < pivot:
                smaller.append(ptr.data)
            else:
                other.append(ptr.data)
            
            ptr = ptr.next
            
        
        smaller.sort()
        other.sort()
        smaller.append(pivot)
        smaller.merge(other)
        self.head = smaller.head
        
    def isCyclic(self):
        ptr1 = self.head
        ptr2 = self.head
        
        while ptr2 != None:
            ptr1 = ptr1.next
            
            if ptr2.next != None:
                ptr2 = ptr2.next.next
            else:
                return False
            
            if ptr1 is ptr2:
                return True
        
        return False

"""
L1 = LinkedList()
L1.appendAll([3,1,2,10])

L2 = LinkedList()
L2.appendAll([5,9,7,2])

L1.merge(L2)
L2.head.data = 23

L1.sort()
print(L1)
"""

ls = LinkedList()
ls.appendAll([3,1,2,10])

print(ls.isCyclic()) 
ls.head.next.next = ls.head 
print(ls.isCyclic()) 
```

```python
class DNode:
    def __init__(self, d, n, p):
        self.data = d
        self.next = n
        self.prev = p

class DLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self.length = 0
    
    def __str__(self):
        st = ""
        ptr = self.head
        while ptr != None:
            st = st + str(ptr.data)
            st = st + " -> "
            ptr = ptr.next
        st += "None and revered: "
        
        ptr = self.tail
        while ptr != None:
            st = st + str(ptr.data)
            st = st + " -> "
            ptr = ptr.prev
        return st + "None"
        
    def search(self, d):
        i = 0
        ptr = self.head
        
        while ptr != None:
            if ptr.data == d:
                return i
            i += 1
            ptr = ptr.next
        return -1
    
    def append(self, d):
        node = DNode(d, None, self.tail)
        self.length += 1
        if not self.tail:  
            self.tail = self.head = node
        else:
            self.tail.next = node
            self.tail = node

    def insert(self, i, d): 
        if i > self.length:
            return
        
        # first element
        if i == 0:
            next = self.head.next
            prev = self.head.prev
            node = DNode(d, next, prev)
            next.prev = node
            self.head = node
            self.length += 1
            return
        
        # last element
        if i == (self.length - 1):
            prev = self.tail.prev
            node = DNode(d, self.tail, prev)
            prev.next = node
            self.tail.prev = node
            self.length += 1
            return
        
        ptr = self.head
        while i > 0 and ptr != None:
            ptr = ptr.next
            i -= 1
            
        prev = ptr.prev
        node = DNode(d, ptr, prev)
        prev.next = node
        ptr.prev = node
        self.length += 1
        
    def remove(self, i):
        if i > self.length:
            return
        
        # first element
        if i == 0:
            val = self.head.data
            self.head = self.head.next
            self.head.prev = None
            self.length -= 1
            return val
        
        # last element
        if i == (self.length - 1):
            val = self.tail.data
            self.tail = self.tail.prev
            self.tail.next = None
            self.length -= 1
            return val
        
        ptr = self.head
        while i > 0 and ptr != None:
            ptr = ptr.next
            i -= 1
            
        ptr.prev.next = ptr.next
        ptr.next.prev = ptr.prev
            
    
DL = DLinkedList()
DL.append(1)
DL.append(2)
DL.append(3)
DL.append(4)
DL.append(5)

print(DL)
DL.remove(2)
print(DL)
```
